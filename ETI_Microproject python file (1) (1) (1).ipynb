```python
# Split a Data set for Testing and Trening purpose

X = df.drop('Temp', axis = 1)
```

```python
y = df.drop('Hum', axis = 1)
```

```python
print("shape of X = ", X.shape)
print("shape of y = ", y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print("shape of X_train = ", X_train.shape )
print("shape of X_test = ", X_test.shape )
print("shape of y_train = ", y_train.shape )
print("shape of y_test = ", y_test.shape )
```

```python
X_train
```

# Linear Regression

```python
# select a model and train it

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
# equation of line
# y = mx + c
```

```python
# value of m(slope)
lr.coef_
```

```python
# value of c (intercept)
lr.intercept_
```

```python
# put value of Humedity to predict a tempreture
lr.predict([[33.0]])
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
# comparision of real data and predicted data

pd.df(np.c_[X_test, y_test, y_pred], columns = ['Hum', 'Temp_original', 'Temp_pred'])
```

```python
# cheack accurasy of model

lr.score(X_test, y_test)
```
